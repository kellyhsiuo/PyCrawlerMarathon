## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

```python
import requests

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
response
```

```python
import json

data = json.loads(response)
for d in data:
    print(d['title'])
```

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

'''
Your Code
'''

print("length of data is", len(data))

for i in data[0].keys():
    print(str(i))
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

'''
Your Code
'''

for row in data:
    print(row['title'],row['createdAt'],row['commentCount'],row['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

'''
Your Code
'''

import requests

r_unpop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false') 
response_unpop = r_unpop.text

import json

data_unpop = json.loads(response_unpop)


count_pop = len(data)
count_unpop = len(data_unpop)

pop_comment = 0
pop_like = 0
unpop_comment = 0
unpop_like = 0

for row in data:
    pop_comment += row['commentCount'] #大於的時候停止
    pop_like += row['likeCount'] 

    
for row in data_unpop:
    unpop_comment += row['commentCount']
    unpop_like += row['likeCount']
    

print("熱門文章的「平均留言人數」:", pop_comment/count_pop)
print("熱門文章的「平均按讚人數」:", pop_like/count_pop)
print("非熱門文章的「平均留言人數」:", unpop_comment/count_unpop)
print("非熱門文章的「平均按讚人數」:", unpop_like/count_unpop)
```
