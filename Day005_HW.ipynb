```python
import json

data = json.loads(response)
for d in data:
    print(d['title'])
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

'''
Your Code
'''

import requests

r_unpop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false') 
response_unpop = r_unpop.text

import json

data_unpop = json.loads(response_unpop)


count_pop = len(data)
count_unpop = len(data_unpop)

pop_comment = 0
pop_like = 0
unpop_comment = 0
unpop_like = 0

for row in data:
    pop_comment += row['commentCount'] #å¤§æ–¼çš„æ™‚å€™åœæ­¢
    pop_like += row['likeCount'] 

    
for row in data_unpop:
    unpop_comment += row['commentCount']
    unpop_like += row['likeCount']
    

print("ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€:", pop_comment/count_pop)
print("ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€:", pop_like/count_pop)
print("éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€:", unpop_comment/count_unpop)
print("éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€:", unpop_like/count_unpop)
```
